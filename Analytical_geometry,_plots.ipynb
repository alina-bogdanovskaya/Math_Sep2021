```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import cm, colors
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import math
from pylab import *
from scipy.optimize import fsolve
```

```python
# Vector length
def vlength(x, y, z):
    a = math.sqrt(x ** 2 + y ** 2 + z ** 2)
    return a

print(f'{vlength(10,10,-10):.2f}')
print(f'{vlength(0,0,-10):.2f}')
```

```python
# Coordinates conversion 
x = float(input('Enter numeric x: '))
y = float(input('Enter numeric y: '))

def cart2pol(x, y):
    R = math.sqrt( x**2 + y**2 )
    alpha = math.atan2(y, x)
    alpha = 180 * alpha/math.pi
    print(f'radius = {R:.2f}, angle = {alpha:.2f}')

cart2pol(x,y)

R = float(input('Enter radius: '))
alpha = float(input('Enter angle in degrees: '))

def pol2cart(R, alpha):
    alpha = alpha * math.pi/180.0
    x = R * math.cos(alpha)
    y = R * math.sin(alpha)
    print(f'x = {x:.2f}, y = {y:.2f}')

pol2cart(R, alpha)
```

```python
x = np.linspace(-5, 5, 201)

plt.rcParams["figure.figsize"] = [10, 7.50]
ax = plt.figure().add_subplot(1, 1, 1)
ax.plot(x, (np.exp(x) - 1)/x + 1)
ax.plot(x, x**2 - 1)

major_ticks_x = np.arange(-6, 6, 1)
minor_ticks_x = np.arange(-6, 6, 0.2)
major_ticks_y = np.arange(-2, 32, 2)
minor_ticks_y = np.arange(-2, 32, 0.4)
ax.set_xticks(major_ticks_x)
ax.set_xticks(minor_ticks_x, minor=True)
ax.set_yticks(major_ticks_y)
ax.set_yticks(minor_ticks_y, minor=True)
ax.grid(which='both')
ax.grid(which='minor', alpha=0.2)
ax.grid(which='major', alpha=0.5)

def equations(p):
  x, y = p
  return (x**2 - 1-y, (np.exp(x) - 1)/x + 1-y)

x0, y0 = fsolve(equations, (0, -1))
print(x0, y0)
x1, y1 = fsolve(equations, (-2, 1))
print(x1, y1)
x2, y2 = fsolve(equations, (2, 6))
print(x2, y2)
x3, y3 = fsolve(equations, (4, 16))
print(x3, y3)
```
