```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1, 10, 121)
k1 = 1
k2 = 2.5
plt.plot(x, np.cos(k1*x), 'c')
plt.plot(x, np.cos(k2*x), 'm')
plt.title('Cosine waves')
plt.xlabel("X")
plt.ylabel("Y")
plt.axhline(y = 0, color = 'k', ls = '--', lw = 0.75)
```
